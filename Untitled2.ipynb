```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Creating the Data Set & Visualization (Pure NumPy) ---

def create_synthetic_dataset(n_samples=500):
    """
    Generates a synthetic binary classification dataset (two Gaussian blobs)
    using only NumPy, adhering to the requirement of labels being 0s and 1s.
    """
    print("1. Generating and Visualizing Synthetic Dataset...")
    np.random.seed(42)

    # Class 0: Data points centered around (1.5, 1.5)
    X0 = np.random.randn(n_samples // 2, 2) * 0.8 + np.array([1.5, 1.5])
    Y0 = np.zeros((1, n_samples // 2))

    # Class 1: Data points centered around (-1.5, -1.5)
    X1 = np.random.randn(n_samples // 2, 2) * 0.8 + np.array([-1.5, -1.5])
    Y1 = np.ones((1, n_samples // 2))

    # Combine data and transpose X to fit convention (features, samples) -> (2, 500)
    X = np.concatenate((X0, X1), axis=0).T
    Y = np.concatenate((Y0, Y1), axis=1)

    # Visualization
    plt.figure(figsize=(8, 6))
    plt.scatter(X[0, :], X[1, :], c=Y[0], cmap=plt.cm.coolwarm, marker='o')
    plt.title("Visualization of Synthetic Data (Class 0 and Class 1)")
    plt.xlabel("Feature X1")
    plt.ylabel("Feature X2")
    plt.show()

    print(f"2. Dataset ready as NumPy arrays: X shape {X.shape}, Y shape {Y.shape}")
    return X, Y

# --- 3. Defining Core Functions (Activation and Initialization) ---

def sigmoid(Z):
    """Sigmoid activation function"""
    A = 1 / (1 + np.exp(-Z))
    return A

def initialize_parameters(layer_dims):
    """4. Initializes the weights (W) and biases (b) for the network."""
    parameters = {}
    L = len(layer_dims) - 1 # Number of layers with parameters (Input is not counted)

    for l in range(1, L + 1):
        # He initialization for ReLU layers: np.sqrt(2/input_dim)
        parameters['W' + str(l)] = np.random.randn(
            layer_dims[l], layer_dims[l-1]) * np.sqrt(2 / layer_dims[l-1])
        parameters['b' + str(l)] = np.zeros((layer_dims[l], 1))

    return parameters

# --- 5. Training the Model (Forward/Backward Propagation in one place) ---

def deep_neural_network_model(X, Y, layer_dims, learning_rate=0.01, num_iterations=15000):
    """
    Main training loop combining Forward Prop, Cost, Back Prop, and Update.
    """
    np.random.seed(1)
    costs = []
    accuracies = [] # List to store accuracy history

    # 4. Initialization
    parameters = initialize_parameters(layer_dims)
    L = len(parameters) // 2 # Number of layers

    print(f"\nTraining Model ({L}-layer NN, Iterations: {num_iterations})")
    print("-" * 40)

    # Loop for Gradient Descent (Training)
    for i in range(0, num_iterations):

        # --- FORWARD PROPAGATION ---
        caches = []
        A = X

        # Layer 1 to L-1 (Hidden Layers: LINEAR -> ReLU)
        for l in range(1, L):
            W = parameters['W' + str(l)]
            b = parameters['b' + str(l)]
            Z = np.dot(W, A) + b
            A_prev = A
            A = np.maximum(0, Z) # ReLU
            caches.append(((A_prev, W, b), Z))

        # Layer L (Output Layer: LINEAR -> Sigmoid)
        W = parameters['W' + str(L)]
        b = parameters['b' + str(L)]
        Z = np.dot(W, A) + b
        AL = sigmoid(Z)
        caches.append(((A, W, b), Z))

        # --- COST CALCULATION ---
        m = Y.shape[1]
        # Adding a small value (1e-8) to log arguments to prevent log(0) errors
        cost = (-1/m) * np.sum(Y * np.log(AL + 1e-8) + (1 - Y) * np.log(1 - AL + 1e-8))
        cost = np.squeeze(cost)

        # Store metrics for plotting
        if i % 100 == 0:
            costs.append(cost)

            # Calculate and store training accuracy
            Y_prediction_train = (AL > 0.5).astype(int)
            current_accuracy = 100 - np.mean(np.abs(Y_prediction_train - Y)) * 100
            accuracies.append(current_accuracy)

            if i % 1000 == 0:
                print(f"Cost after iteration {i:5}: {cost:.4f} | Accuracy: {current_accuracy:.2f}%")

        # --- BACKWARD PROPAGATION ---
        grads = {}
        dAL = - (np.divide(Y, AL) - np.divide(1 - Y, 1 - AL)) # Initial derivative

        # Output Layer (Sigmoid Back)
        current_cache = caches[L-1]
        ((A_prev, W, b), Z) = current_cache

        dZ = dAL * sigmoid(Z) * (1 - sigmoid(Z)) # Sigmoid derivative
        dW = (1/m) * np.dot(dZ, A_prev.T)
        db = (1/m) * np.sum(dZ, axis=1, keepdims=True)
        dA_prev = np.dot(W.T, dZ)

        grads["dW" + str(L)] = dW
        grads["db" + str(L)] = db

        # Hidden Layers (ReLU Back)
        for l in reversed(range(L - 1)):
            current_cache = caches[l]
            ((A_prev, W, b), Z) = current_cache

            dZ = np.array(dA_prev, copy=True)
            dZ[Z <= 0] = 0 # ReLU derivative

            dW = (1/m) * np.dot(dZ, A_prev.T)
            db = (1/m) * np.sum(dZ, axis=1, keepdims=True)
            dA_prev = np.dot(W.T, dZ)

            grads["dW" + str(l + 1)] = dW
            grads["db" + str(l + 1)] = db

        # --- UPDATE PARAMETERS ---
        for l in range(1, L + 1):
            parameters["W" + str(l)] -= learning_rate * grads["dW" + str(l)]
            parameters["b" + str(l)] -= learning_rate * grads["db" + str(l)]

    print("-" * 40)

    # Calculate Final Accuracy (based on the last iteration's results)
    Y_prediction_train = (AL > 0.5).astype(int)
    final_accuracy = 100 - np.mean(np.abs(Y_prediction_train - Y)) * 100
    print(f"Final Training Accuracy: {final_accuracy:.2f}%")

    # Plot the loss and accuracy side-by-side
    plt.figure(figsize=(15, 6))

    # Subplot 1: Loss
    plt.subplot(1, 2, 1)
    plt.plot(costs)
    plt.ylabel('Cost (Loss)')
    plt.xlabel('Epochs (x100)')
    plt.title("1. Loss over Training Epochs (Should decrease)")
    plt.grid(True)

    # Subplot 2: Accuracy
    plt.subplot(1, 2, 2)
    plt.plot(accuracies, color='green')
    plt.ylabel('Accuracy (%)')
    plt.xlabel('Epochs (x100)')
    plt.title("2. Accuracy over Training Epochs (Should increase)")
    plt.grid(True)

    plt.tight_layout()
    plt.show()

# --- Main Execution ---

if __name__ == "__main__":

    # 1 & 2. Create and prepare data
    X_train, Y_train = create_synthetic_dataset(n_samples=500)

    # 3. Define the architecture: Input(2) -> Hidden(4) -> Output(1)
    # This architecture defines two weight matrices (W1, W2) for the 3-layer NN
    layer_dims = [X_train.shape[0], 4, 1]

    # 5. Train the model and plot
    deep_neural_network_model(
        X_train, Y_train,
        layer_dims,
        learning_rate = 0.01,
        num_iterations = 15000
    )
```
